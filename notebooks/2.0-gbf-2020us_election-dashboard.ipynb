```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime, timedelta

from statsmodels.stats.weightstats import DescrStatsW

import us

import sys

import os
path = os.getcwd()
path = path[:path.find('us_election2020')] + 'us_election2020'

sys.path.append(path + '\\src\\data')
import make_dataset
```

```python
import plotly.offline as py
import plotly.graph_objs as go
```

---

```python
election_polls2020 = pd.read_csv(path+'\\data\\processed\\general_election_polls2020.csv', parse_dates=True, index_col=0)
```

```python
election_polls2020 = election_polls2020[election_polls2020['start_date']>'2020-01-01']
```

```python
national_stats_k16 = pd.read_csv(path+'\\data\\processed\\2020_National_stats_at16.csv', parse_dates=True, index_col=0).loc['2020-01-15':]
```

```python
line_biden = go.Scatter(mode='lines',
            x=national_stats_k16.index, 
            y=national_stats_k16['Biden_avg'], 
            name='Biden',
            line={'color':'blue'})
fill_biden = go.Scatter(mode='lines',
            x=national_stats_k16.index.tolist()+national_stats_k16.index[::-1].tolist(), 
            y=(national_stats_k16['Biden_avg']+national_stats_k16['Biden_std']).tolist()+(national_stats_k16['Biden_avg']-national_stats_k16['Biden_std'])[::-1].tolist(), 
            line={'color':'rgba(255,255,255,0)'}, 
            fillcolor='rgba(0,0,,0.2)', fill='toself', hoverinfo="skip", showlegend=False)


scatter_biden = go.Scatter(
            mode='markers',
            marker=dict(color='blue', size=2),
            x=election_polls2020['start_date'], 
            y=election_polls2020['Biden'], 
            opacity=0.5, 
            name='Biden polls', showlegend=False
        )
text_biden = go.Scatter(
            mode='text',
            x=[national_stats_k16.index[-1].date() + timedelta(days=10.5)], 
            y=[national_stats_k16['Biden_avg'].iloc[-1]], 
            text=[str(round(national_stats_k16['Biden_avg'].iloc[-1], 1))], 
            name='Biden polls', showlegend=False
        )


trace_trump = go.Scatter(mode='lines',
            x=national_stats_k16.index, 
            y=national_stats_k16['Trump_avg'], 
            name='Trump',
            line={'color':'red'}
        )
scatter_trump = go.Scatter(
            mode='markers',
            marker=dict(color='red', size=2),
            x=election_polls2020['start_date'], 
            y=election_polls2020['Trump'], 
            opacity=0.5, 
            name='Trump polls', showlegend=False
        )
text_trump = go.Scatter(
            mode='text',
            x=[national_stats_k16.index[-1].date() + timedelta(days=10.5)], 
            y=[national_stats_k16['Trump_avg'].iloc[-1]], 
            text=[str(round(national_stats_k16['Trump_avg'].iloc[-1], 1))], 
            name='Trump polls', showlegend=False
        )

data = [line_biden, scatter_biden, text_biden, fill_biden, 
        trace_trump, scatter_trump, text_trump]

layout = {
    'title':'2020 US General Election', 
     'yaxis':{'autorange':False, 
              'range':[35, 55],
              'type':"linear"}
}

fig = {'data':data, 'layout':layout}
py.iplot(fig)
```
